# Merging embeddings and clinical data
This notebook serves merging image embeddings with clinical data acquired by Hinrich B. Winther et al. (2020, [DOI: 10.6084/m9.figshare.12275009](https://doi.org/10.6084/m9.figshare.12275009)).

Results are stored as data_merged.csv and can be used in VEST to visualize distribution of clinical parameters in 3D space.

```python
import pandas as pd

# Load the first CSV file
df1 = pd.read_csv('data.csv')
print("Columns in data.csv:")
print(df1.columns.tolist())
print()

# Load the second CSV file
df2 = pd.read_csv('covid-19-image-repository/data.csv')
print("Columns in covid-19-image-repository/data.csv:")
print(df2.columns.tolist())
```

```python
df1["filename"][0], df2["image_id"][0]
```

```python
df1.shape, df2.shape
```

```python
# Extract patient_id from filename by removing the .png extension
df1['image_id'] = df1['filename'].str.replace('.png', '')

# Join the dataframes on patient_id
merged_df = pd.merge(df1, df2, on='image_id', how='inner')

print(f"df1 shape: {df1.shape}")
print(f"df2 shape: {df2.shape}")
print(f"Merged dataframe shape: {merged_df.shape}")
print("\nMerged dataframe columns:")
print(merged_df.columns.tolist())

merged_df.head()
```

```python
# Save the merged dataframe to CSV
merged_df.to_csv('data_merged.csv', index=False)
print("merged_df saved as data_merged.csv")
```
