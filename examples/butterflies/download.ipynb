# Download Butterfly images

This notebooks downloads a dataset of [butterfly images](https://www.kaggle.com/datasets/phucthaiv02/butterfly-image-classification) published on kaggle by DePie.

In order to run it, you need to install [kagglehub](https://github.com/Kaggle/kagglehub):

```
pip install kagglehub
```


````python
import os
import kagglehub

os.makedirs("./images", exist_ok=True)

# Download latest version
path = kagglehub.dataset_download("phucthaiv02/butterfly-image-classification", output_dir="./images")

print("Path to dataset files:", path)
````
